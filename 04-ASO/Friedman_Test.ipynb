# Calculo Friedman Test

```python
# Seleccionar las columnas 'MSE', 'RMSE', 'MAE', 'R2', 'cnSMAPE' del DataFrame
g1 = df['MSE'].values
g2 = df['RMSE'].values
g3 = df['MAE'].values
g4 = df['R2'].values
g5 = df['cnSMAPE'].values

# Realizar la prueba de Friedman
Friedman_T = stats.friedmanchisquare(g1, g2, g3, g4, g5)

# Mostrar los resultados de la prueba de Friedman
print('Estadístico de la prueba de Friedman:', Friedman_T.statistic)
print('Valor p de la prueba de Friedman:', Friedman_T.pvalue)
```

```python
# Eliminar las primeras dos filas del DataFrame
df.drop(df.index[:2], inplace=True)

# Seleccionar las columnas 'MSE', 'RMSE', 'MAE', 'R2', 'cnSMAPE' del DataFrame modificado
g1 = df['MSE'].values
g2 = df['RMSE'].values
g3 = df['MAE'].values
g4 = df['R2'].values
g5 = df['cnSMAPE'].values

# Realizar la prueba de Friedman
Friedman_T = stats.friedmanchisquare(g1, g2, g3, g4, g5)

# Mostrar los resultados de la prueba de Friedman
print('Estadístico de la prueba de Friedman:', Friedman_T.statistic)
print('Valor p de la prueba de Friedman:', Friedman_T.pvalue)
```
