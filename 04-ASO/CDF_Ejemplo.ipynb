```python
# Importar las bibliotecas necesarias
import numpy as np  # Importar NumPy para cálculos numéricos
import matplotlib.pyplot as plt  # Importar Matplotlib para trazar gráficos
import pandas as pd  # Importar pandas para manejo de datos en forma de tablas
```

```python
# Paso 2: Crear o cargar la serie de tiempo (aquí se proporciona un ejemplo)
serie_tiempo = [10, 15, 12, 20, 18, 25, 22, 30, 28, 35]

# Paso 3: Calcular la función de distribución acumulada (CDF)
# Utilizaremos la función `numpy` `cumsum` para obtener la suma acumulada.
cdf = np.cumsum(serie_tiempo)

# Paso 4: Graficar la CDF
plt.plot(cdf, marker='o')  # Graficar la CDF con marcadores circulares en los puntos
plt.xlabel('Índice')  # Etiqueta del eje x
plt.ylabel('CDF')  # Etiqueta del eje y
plt.title('Función de Distribución Acumulada (CDF) de la Serie de Tiempo')  # Título del gráfico
plt.grid(True)  # Habilitar la cuadrícula en el gráfico
plt.show()  # Mostrar el gráfico
```

```python
# Ejemplo de Serie (puedes reemplazar esto con tus datos)
data = pd.Series([1, 2, 2, 3, 3, 3, 4, 4, 5])
# data = pd.Series([10, 15, 12, 20, 18, 25, 22, 30, 28, 35])

# Calcular la CDF
# Utilizamos value_counts() para contar las ocurrencias de cada valor, luego normalizamos y calculamos la suma acumulada.
cdf = data.value_counts(normalize=True).sort_index().cumsum()

# Graficar la CDF
plt.plot(cdf.index, cdf.values, marker='o')  # Graficar la CDF con marcadores circulares en los puntos
plt.xlabel('Valor')  # Etiqueta del eje x
plt.ylabel('CDF')  # Etiqueta del eje y
plt.title('Función de Distribución Acumulada (CDF)')  # Título del gráfico
plt.grid(True)  # Habilitar la cuadrícula en el gráfico
plt.show()  # Mostrar el gráfico
```

```python
# Crear una serie de números aleatorios
data = np.random.rand(100)

# Calcular la CDF
# Calculamos la suma acumulada y la dividimos por la longitud de la serie para obtener la función de distribución acumulada (CDF).
cdf = np.cumsum(data) / len(data)

# Graficar la CDF
plt.plot(cdf)  # Graficar la CDF
plt.show()  # Mostrar el gráfico
```

```python
# Convertir los datos en una Serie de Pandas
data = pd.Series(data)

# Calcular la CDF
# Calculamos la frecuencia relativa de cada valor en la serie, la ordenamos por índice y luego calculamos la suma acumulada.
cdf = data.value_counts(normalize=True).sort_index().cumsum()

# Graficar la CDF
plt.plot(cdf.index, cdf.values, marker='o')  # Graficar la CDF
plt.xlabel('Value')  # Etiqueta del eje x
plt.ylabel('CDF')  # Etiqueta del eje y
plt.title('Cumulative Distribution Function (CDF)')  # Título del gráfico
plt.grid(True)  # Mostrar cuadrícula en el gráfico
plt.show()  # Mostrar el gráfico
```

```python
# Calcular la suma acumulada de los valores en la serie
cdf = np.cumsum(data.values)

# Graficar la CDF
plt.plot(cdf, marker='o')  # Graficar la CDF
plt.xlabel('Índice')  # Etiqueta del eje x
plt.ylabel('CDF')  # Etiqueta del eje y
plt.title('Función de Distribución Acumulada (CDF) de la Serie de Tiempo')  # Título del gráfico
plt.grid(True)  # Mostrar cuadrícula en el gráfico
plt.show()  # Mostrar el gráfico
```
