# Calculo Friedman Test

```python
g1=df['MSE'].values
g2=df['RMSE'].values
g3=df['MAE'].values
g4=df['R2'].values
g5=df['cnSMAPE'].values

#perform Friedman Test
Friedman_T = stats.friedmanchisquare( g1, g2, g3, g4, g5 )

print('Static: ', Friedman_T.statistic)
print('pvalue: ', Friedman_T.pvalue)
```
